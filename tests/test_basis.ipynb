```python
import sys
sys.path.append('..')

import torch
import matplotlib.pyplot as plt
from equitorch.nn._radial_basis import GaussianBasisExpansion, BesselBasisExpansion
from equitorch.nn._angular_basis import FourierBasisExpansion, SineBasisExpansion, CosineBasisExpansion
```

```python
N = 100
X = torch.rand(N, N) * 15 - 2.5
gbe = GaussianBasisExpansion(gamma=1, num_basis=10, start=1, end=10)
Y = gbe(X)
Y.shape
```

```python
plt.scatter(X.reshape(N*N,1).repeat_interleave(10,1), Y.flatten(0,1), )
```

```python
N = 100
X = torch.rand(N, N)
bbe = BesselBasisExpansion(num_basis=5, trainable=False)
Y = bbe(X)
Y.shape
```

```python
plt.scatter(X.reshape(N*N,1).repeat_interleave(5,1), Y.flatten(0,1), )
```

```python
N = 100
X = torch.rand(N, N) * torch.pi * 2 - torch.pi
fbe = FourierBasisExpansion(max_freq=2)
Y = fbe(X)
print(Y.shape)
plt.scatter(X.reshape(N*N,1).repeat_interleave(4,1), Y.flatten(0,1), )
```

```python
N = 100
X = torch.rand(N, N) * torch.pi * 2 - torch.pi
fbe = FourierBasisExpansion(max_freq=2, include_freq_0=True)
Y = fbe(X)
print(Y.shape)
plt.scatter(X.reshape(N*N,1).repeat_interleave(5,1), Y.flatten(0,1), )
```
